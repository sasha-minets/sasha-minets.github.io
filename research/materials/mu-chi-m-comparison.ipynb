```python
R.<t> = QQ[]

def qnum(a, b): #quantum number [a]_{t^b}
    res = 0
    s = 1
    for i in range(a):
        res += s
        s *= t^b
    return res

def ispos(p): #checks a sufficient (not necessary!) condition for p to be positive for t>=1; 
    coef = p.list()
    coef.reverse()
    sum = 0
    res = True
    if (p == 0):
        return False
    for c in coef:
        sum += c
        if (sum < 0):
            res = False
    return res

def comp(p,q):
    return ispos(q-p)

def munum(e,w): #denominator of mu is computed from the weights of the base
    return qnum(e,1)*qnum(2*e,1)

def mu11den(e,w): #numerator, partition (1,1)
    return qnum(w,1)*qnum(2*w,1)

def mu2den(e,w): #numerator, partition (2)
    return qnum(w,1)*qnum(2*w-1,1)

def G2(e,p): #function in the definition of some χ's; only works for polynomials with coefficients 0 and 1!!
    coef = p.list()
    k = len(coef)
    s = 0
    for i in range(k):
        for j in range(i+1,k):
            s += coef[i]*coef[j]*t^(i+j)
    return (1+t^e)*s + p.subs(t=t^2)
```
